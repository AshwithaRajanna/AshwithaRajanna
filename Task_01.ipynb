# **Task-01**

**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_data= pd.read_csv("/content/HeartDiseaseTrain-Test.csv")
```

```python
heart_data.head(11)
```

**Categorical variables**:- Categorical variables represent distinct categories or groups.

**Continuous variables:-** Continuous variables represent numerical measurements that can take on any value within a certain range.

```python
# Create a bar chart for the categorical variables
categorical_variables = ['sex', 'chest_pain_type', 'fasting_blood_sugar', 'rest_ecg', 'exercise_induced_angina', 'slope', 'vessels_colored_by_flourosopy', 'thalassemia', 'target']
for variable in categorical_variables:
    value_counts = heart_data[variable].value_counts()
    plt.figure()
    plt.bar(value_counts.index, value_counts.values, color='skyblue')
    plt.xlabel(variable)
    plt.ylabel('Count')
    plt.title(f'Distribution of {variable} in Heart Disease Dataset')


# Create histograms for the continuous variables
continuous_variables = ['age', 'resting_blood_pressure', 'cholestoral', 'Max_heart_rate']
for variable in continuous_variables:
    plt.figure()
    plt.hist(heart_data[variable], bins=10, color='lightgreen', edgecolor='black')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {variable} in Heart Disease Dataset')

plt.show()
```

```python
# Creating a histogram to visualize the distribution of the 'age' variable
plt.hist(heart_data['age'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in Heart Disease Dataset')
plt.show()
```

```python
# Creating a bar chart to visualize the distribution of the 'sex' variable
sex_counts = heart_data['sex'].value_counts()
plt.bar(sex_counts.index, sex_counts.values, color=['skyblue', 'salmon'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in Heart Disease Dataset')
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()
```
